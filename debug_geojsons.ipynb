```python
import folium
import os
import json

files = os.listdir('debug')

files.sort()

for file in files:
    with open('debug/' + file, 'r') as f:
        geojson = json.load(f)
    
    f = folium.Figure(width=1000, height=500)
    m = folium.Map(location=[51.050407, 13.737262], zoom_start=13,).add_to(f)
    folium.TileLayer('cartodbdark_matter').add_to(m)
    
    folium.GeoJson(geojson,
                   style_function=lambda feature: {
                        "color": "green"
                        if feature["properties"]["type"] == "track"
                        else "blue" if feature["properties"]["type"] == "map_matched_track" else "gray",
                        "weight": 2 if feature["properties"]["type"] == "track" or feature["properties"]["type"] == "map_matched_track" else 0.5,
                    }).add_to(m)
    
    title_html = f"""
        <h3> {file} </h3>
    """
    m.get_root().html.add_child(folium.Element(title_html))
    
    display(m)   
```
